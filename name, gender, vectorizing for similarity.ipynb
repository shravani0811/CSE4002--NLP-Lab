```python
import nltk
from nltk.corpus import stopwords
```

```python
def gender_features(word):
    return{'last_letter':word[-1]}
```

```python
from nltk.corpus import names
```

```python
names.words()
```

```python
len(names.words())
```

```python
labeled_names=[(name, 'male') for name in names.words('male.txt')]+[(name,'female') for name in names.words('female.txt')]
```

```python
labeled_names
```

```python
import random
random.shuffle(labeled_names)
```

```python
labeled_names
```

```python
featureset=[(gender_features(n),gender) for (n,gender) in labeled_names]
```

```python
featureset
```

```python
train_set, test_set = featureset[:5500],featureset[5501:]
```

```python
test_set
```

```python
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_features('Aradhita'))
```

```python
classifier.classify(gender_features('Obama'))
```

```python
classifier.classify(gender_features('Nitish'))
```

```python
classifier.classify(gender_features('Anurag'))
```

```python
nltk.classify.accuracy(classifier,test_set)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec = CountVectorizer(binary=True)
```

```python
corpus = ["Jack and Jill went up the hill", "Jack fell down and broke his crown"]
```

```python
vec.fit(corpus)
```

```python
voc = vec.vocabulary_
```

```python
voc
```

```python
vec.transform(["The crown rested heavy on his young head"]).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similar = cosine_similarity(vec.transform(["Obama was the 44th president of USA, he was in office for two terms"]).toarray(), vec.transform(["Trump is the 45th president of USA, he has been in office for one term and is standing for reelection"]).toarray())
print(similar)
```
